# Band composition display
* With rasterIO, we can merge the images into color compositions such as RGB by placing a raster to represent the Landsat 8 red, green and blue spectra.


Below we will show the main colored band compositions

````
blue = rio.open('LC08_L1TP_228071_20190816_20190902_01_T1_B2.TIF')
green = rio.open('LC08_L1TP_228071_20190816_20190902_01_T1_B3.TIF')
red = rio.open('LC08_L1TP_228071_20190816_20190902_01_T1_B4.TIF')
nir = rio.open('LC08_L1TP_228071_20190816_20190902_01_T1_B5.TIF')
swir1 = rio.open('LC08_L1TP_228071_20190816_20190902_01_T1_B6.TIF')
swir2 = rio.open('LC08_L1TP_228071_20190816_20190902_01_T1_B7.TIF')
````

Importing function to merge:

````
from rasterio.merge import merge
````

Merge from read files

````
src_files_to_mosaic = [red, green, blue]
mosaic, out_trans = merge(src_files_to_mosaic)
````

* B4(Red), B3(Green) and B2(Blue)

````
show(mosaic, cmap='terrain')
````

Thus we can create various compositions to visualize the data of interest.

* B6(SWIR 1 or SWIR short), B5(NIR) and B4(Red)

````
src_files_to_mosaic = [swir1, nir, red]
mosaic, out_trans = merge(src_files_to_mosaic)
````

````
show(mosaic, cmap='Reds')
````

Images are always represented by numbers, but to be located in geographic space, they need a spatial reference system known as projection.

Plotted image without projection, only numbers:

````
show(red.read())
````

In this example, we can see that the axes of the figure start at 0.


Projected plotted image, numbers + location (meters)

````
show(red)
````

In this example, we can see that the axes of the figure start at a measurement in meters, representing the zones of the UTM system, which is used for ground surface location.


With GDAL, we can see the image projection in detail:

````
!gdalinfo LC08_L1TP_228071_20190816_20190902_01_T1_B2.TIF
````

# gdalinfo identify projection of image
Projections can be represented by codes called EPSG. In this gdalinfo result, the projection is the last result of the AUTHORITY tag:


```
AUTHORITY ["EPSG", "32621"]]
```

This is the projection of the downloaded image. More details about this EPSG at [EPSG: 32621](https://spatialreference.org/ref/epsg/32621/)

# plot raster and vector(distint projection)


Now, let's do the same plot except using the image with a different projection than geojson.

````
vector = gpd.read_file('myData/vector_epsg4326.geojson')
fig, ax = plt.subplots(figsize=(12, 10))
show(red, ax=ax)
vector.buffer(30).plot(ax=ax, edgecolor='red', facecolor='none', linewidth=30)
plt.show()
````

In the example above we can see the huge displacement between the 2 dice, since they are in different projections, the unit of measurements are not the same. This vector of the example has its unit of measure in degrees, and the raster in meters. Thus causing the vector to be plotted near 0.

# Homework

Export one of the created 3 band compositions to a single TIF file.

# Homework solution

````
with rio.open('LC08_L1TP_228071_20190816_20190902_01_T1_B4.TIF') as src:
  profile = src.profile.copy()
  red = src.read(1)
with rio.open('LC08_L1TP_228071_20190816_20190902_01_T1_B3.TIF') as src:
  green = src.read(1)
with rio.open('LC08_L1TP_228071_20190816_20190902_01_T1_B2.TIF') as src:
  blue = src.read(1)

profile_out = profile.copy()
profile_out['count'] = 3
````

````
profile_out
````

````
with rio.open('rgb.tif', 'w', **profile_out) as rgb:
  rgb.write(red, 1)
  rgb.write(green, 2)
  rgb.write(blue, 3)
````
